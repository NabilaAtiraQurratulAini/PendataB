**Langkah 1 : Mengimpor Perpustakaan**

Seperti biasa, langkah pertama akan selalu mencakup mengimpor perpustakaan yaitu NumPy, Pandas dan Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

**Langkah 2 : Mengimpor himpunan data**

Pada langkah ini, kita akan mengimpor dataset Iris Flower yang disimpan di repositori github saya sebagai dan menyimpannya ke variabel Setelah ini, kami menetapkan 4 variabel independen ke X dan variabel dependen 'spesies' ke Y. 5 baris pertama himpunan data ditampilkan.IrisDataset.csvdataset.

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/mk-gurucharan/Classification/master/IrisDataset.csv')
X = dataset.iloc[:,:4].values
y = dataset['species'].values
dataset.head(5)
```

**Langkah 3 : Memisahkan himpunan data ke dalam Set pelatihan dan Set pengujian**

Setelah kami memperoleh kumpulan data kami, kami harus membagi data menjadi set pelatihan dan set pengujian. Dalam kumpulan data ini, ada 150 baris dengan 50 baris dari masing-masing 3 kelas. Karena setiap kelas diberikan dalam urutan kontinu, kita perlu membagi himpunan data secara acak. Di sini, kami memiliki , yang berarti bahwa 20% dari himpunan data akan digunakan untuk tujuan pengujian sebagai set pengujian dan 80% sisanya akan digunakan sebagai set pelatihan untuk melatih model klasifikasi Naive Bayes.test_size=0.2

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

**Langkah 4 : Penskalaan Fitur**

Himpunan data diperkecil ke rentang yang lebih kecil menggunakan opsi Penskalaan Fitur. Dalam hal ini, nilai dan diperkecil ke nilai yang lebih kecil untuk meningkatkan kecepatan program.X_trainX_test

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

**Langkah 5 : Melatih model Klasifikasi Naive Bayes pada Set Pelatihan**

Pada langkah ini, kami memperkenalkan kelas yang digunakan dari perpustakaan. Di sini, kami telah menggunakan model Gaussian, ada beberapa model lain seperti Bernoulli, Categorical dan Multinomial. Di sini, kami menetapkan kelas GaussianNB ke variabel dan sesuai dengan nilai X_train dan y_train untuk tujuan pelatihan.GaussianNBsklearn.naive_bayesclassifier

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

**Langkah 6 : Memprediksi hasil set Tes**

Setelah model dilatih, kita menggunakan untuk memprediksi nilai untuk set Uji dan nilai yang diprediksi disimpan ke variabel classifier.predict()y_pred.

```python
y_pred = classifier.predict(X_test) 
y_pred
```

**Langkah 7 : Matriks Kebingungan dan Akurasi**

Ini adalah langkah yang banyak digunakan dalam teknik klasifikasi. Dalam hal ini, kita melihat Akurasi model terlatih dan memplot matriks kebingungan.

Matriks kebingungan adalah tabel yang digunakan untuk menunjukkan jumlah prediksi yang benar dan salah pada masalah klasifikasi ketika nilai sebenarnya dari Set Tes diketahui. Ini dari format.

Nilai Sebenarnya adalah jumlah prediksi yang benar yang dibuat.

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
from sklearn.metrics import accuracy_score 
print ("Accuracy : ", accuracy_score(y_test, y_pred))
cm
```

Dari matriks kebingungan di atas, kami menyimpulkan bahwa, dari 30 data set pengujian, 29 diklasifikasikan dengan benar dan hanya 1 yang salah diklasifikasikan. Ini memberi kami akurasi tinggi 96,67%.

**Langkah 8: Membandingkan Nilai Riil dengan Nilai Prediksi**

Pada langkah ini, Pandas DataFrame dibuat untuk membandingkan nilai yang diklasifikasikan dari set Uji asli (y_test) dan hasil prediksi (y_pred).

```python
df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
df
```

Langkah ini adalah langkah tambahan yang tidak terlalu informatif seperti matriks Kebingungan dan terutama digunakan dalam regresi untuk memeriksa keakuratan nilai yang diprediksi.

Seperti yang Anda lihat, ada satu prediksi yang salah yang telah memprediksi versicolor, bukan virginica.
