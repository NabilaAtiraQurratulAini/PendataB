### **PYTHON DECISION TREE DATA WEATHER**

**Mulai Pemrograman**

Hal pertama yang harus dilakukan adalah menginstal dependensi atau pustaka yang akan membuat program ini lebih mudah untuk ditulis. Saya akan mengimpor perpustakaan pembelajaran mesin sklearn, pandas, pydontplus dan IPython.display.

```python
## import dependencies
from sklearn import tree #For our Decision Tree
import pandas as pd # For our DataFrame
import pydotplus # To create our Decision Tree Graph
from IPython.display import Image  # To Display a image of our graph
```

Selanjutnya akan membuat kumpulan data yang akan digunakan untuk contoh ini pada Pohon Keputusan, dengan terlebih dahulu membuat bingkai data pandas kosong, dan memasukkan data ke setiap kolom / fitur / atribut (Outlook, Temperature, Humidity, Windy, Play).

```python
#Create the dataset
#create empty data frame
golf_df = pd.DataFrame()

#add outlook
golf_df['Outlook'] = ['sunny', 'sunny', 'overcast', 'rainy', 'rainy', 'rainy', 'overcast', 'sunny', 'sunny', 'rainy', 'sunny', 'overcast', 'overcast', 'rainy']

#add temperature
golf_df['Temperature'] = ['hot', 'hot', 'hot', 'mild', 'cool', 'cool', 'cool', 'mild', 'cool', 'mild', 'mild', 'mild', 'hot', 'mild']

#add humidity
golf_df['Humidity'] = ['high', 'high', 'high', 'high', 'normal', 'normal', 'normal', 'high', 'normal', 'normal', 'normal', 'high', 'normal', 'high']

#add windy
golf_df['Windy'] = ['false', 'true', 'false', 'false', 'false', 'true', 'true', 'false', 'false', 'false', 'true', 'true', 'false', 'true']

#finally add play
golf_df['Play'] = ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']


#Print/show the new data
print(golf_df)
```

Sekarang akan mengubah variabel kategoris (Outlook, Temperature, Humidity, Windy, Play) menjadi variabel dummy / indikator atau (variabel biner) pada dasarnya 1 dan 0.

```python
# Convert categorical variable into dummy/indicator variables or (binary vairbles) essentialy 1's and 0's
# I chose the variable name one_hot_data bescause in ML one-hot is a group of bits among which the legal combinations of values are only those with a single high (1) bit and all the others low (0)
one_hot_data = pd.get_dummies(golf_df[ ['Outlook', 'Temperature', 'Humidity', 'Windy'] ])

#print the new dummy data
one_hot_data
```

Akhirnya sampai pada titik membuat dan melatih Pengklasifikasi Pohon Keputusan (model)!

```python
# The decision tree classifier.
clf = tree.DecisionTreeClassifier()

# Training the Decision Tree
clf_train = clf.fit(one_hot_data, golf_df['Play'])
```

Selanjutnya akan membuat grafik Pohon Keputusan untuk mendapatkan visual yang lebih baik dari apa yang dilakukan model, dengan mencetak data DOT pohon, membuat grafik data DOT menggunakan metode pydontplus graph_from_dat_data dan menampilkan grafik menggunakan metode Gambar IPython.display.

```python
# Export/Print a decision tree in DOT format.
print(tree.export_graphviz(clf_train, None))

#Create Dot Data
dot_data = tree.export_graphviz(clf_train, out_file=None, feature_names=list(one_hot_data.columns.values), class_names=['Not_Play', 'Play'], rounded=True, filled=True)

#Gini decides which attribute/feature should be placed at the root node, which features will act as internal nodes or leaf nodes
#Create Graph from DOT data
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```

Terakhir namun tidak kalah pentingnya, buat prediksi, dengan memasukkan Outlook sebagai 'cerah', Suhu sebagai 'panas', Kelembaban sebagai 'normal' dan Windy sebagai 'salah'. Model memperkirakan masukan itu adalah 'ya', artinya pegolf akan bermain golf hari itu. Program selesai! Tentu saja ini hanyalah contoh sederhana dari Pengklasifikasi Pohon Keputusan. Pada kenyataannya Anda kemungkinan besar akan mengunggah kumpulan data alih-alih membuatnya, lalu membersihkan dan menjelajahi data, lalu membagi data Anda menjadi kumpulan pelatihan dan rangkaian pengujian dan menguji akurasi model Anda menggunakan beberapa metrik statistik.
